## 2. Döngülər & Şərtlər

```python
# Siyahı üzərindən döngü (Iterate over a list)
meyveler = ["alma", "gilas", "nar"]
for meyve in meyveler:
    print(meyve)
```

```python
# Range ilə döngü (Range-based loop)
for i in range(5):
    print(i)  # 0, 1, 2, 3, 4
```

```python
# Şərtlər (Conditionals)
nətice = 85

if nətice >= 90:
    qiymet = "A"
elif nətice >= 80:
    qiymet = "B"
else:
    qiymet = "C"

print(f"Nəticə: {nətice}, Qiymət: {qiymet}")
```

```python
# While döngüsü
say = 0
while say < 3:
    print(f"Say: {say}")
    say += 1
```

## 3. Funksiyalar

```python
def salamlama(ad):
    return f"Salam, {ad}!"

mesaj = salamlama("PyWeek")
print(mesaj)
```

```python
# Daha mürəkkəb funksiya nümunəsi
def find_max(numbers):
    """Siyahıda ən böyük ədədi tapır (max() istifadə etmədən)."""
    if not numbers:
        return None
    biggest = numbers[0]
    for num in numbers:
        if num > biggest:
            biggest = num
    return biggest

test_list = [42, 17, 93, 8, 55]
print(f"Ən böyük: {find_max(test_list)}")
```

## 4. Lüğətlər (Dictionaries) — Açar-Dəyər Cütlükləri

```python
# Lüğət yarat (Create a dictionary)
gen = {
    "ad": "BRCA1",
    "xromosom": 17,
    "uzunluq": 81189
}

# Dəyərə müraciət (Access values)
print(gen["ad"])  # BRCA1

# Əlavə et / Yenilə (Add / update)
gen["funksiyasi"] = "DNA repair"

# Açarlar və dəyərlər üzərindən döngü
for key, val in gen.items():
    print(f"{key}: {val}")
```

```python
# Lüğət ilə saymaq
text = "bioinformatics"
letter_counts = {}
for char in text:
    letter_counts[char] = letter_counts.get(char, 0) + 1

print("Hərf tezlikləri:")
for letter, count in sorted(letter_counts.items()):
    print(f"  '{letter}': {count}")
```

```python
import csv

# Nümunə CSV yaratmaq
sample_data = [
    {"species": "Canis lupus", "count": "12", "region": "Şimal"},
    {"species": "Aquila chrysaetos", "count": "8", "region": "Dağlıq"},
    {"species": "Cervus elaphus", "count": "23", "region": "Meşə"},
    {"species": "Ursus arctos", "count": "5", "region": "Dağlıq"},
    {"species": "Vulpes vulpes", "count": "31", "region": "Düzənlik"},
    {"species": "Aquila chrysaetos", "count": "3", "region": "Şimal"},
    {"species": "Canis lupus", "count": "7", "region": "Meşə"},
]

with open("observations.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["species", "count", "region"])
    writer.writeheader()
    writer.writerows(sample_data)

print("observations.csv yaradıldı!")
```

```python
# CSV faylını oxumaq
with open("observations.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['species']}: {row['count']} ({row['region']})")
```

```python
# Hamısını birləşdirək — Dataset analizi
import csv

counts = {}  # növləri saymaq üçün lüğət
with open("observations.csv") as f:
    for row in csv.DictReader(f):
        species = row["species"]
        counts[species] = counts.get(species, 0) + int(row["count"])

print("Növ üzrə cəm müşahidə:")
for species, total in sorted(counts.items(), key=lambda x: -x[1]):
    print(f"  {species}: {total}")
```

## 6. NumPy — Ədədi Hesablamalar

```python
import numpy as np

# Massiv yaratmaq
a = np.array([1, 2, 3, 4, 5])
b = np.zeros(10)
c = np.linspace(0, 1, 100)

print(f"a = {a}")
print(f"b = {b}")
print(f"c (ilk 10): {c[:10]}")
```

```python
# Əməliyyatlar (elementwise)
print(f"a * 2  = {a * 2}")     # [2, 4, 6, 8, 10]
print(f"a ** 2 = {a ** 2}")    # [1, 4, 9, 16, 25]
print(f"a + 10 = {a + 10}")    # [11, 12, 13, 14, 15]
```

```python
# Statistika
print(f"Orta (mean): {np.mean(a)}")   # 3.0
print(f"Std:         {np.std(a):.2f}") # 1.41
print(f"Maks:        {np.max(a)}")     # 5
print(f"Min:         {np.min(a)}")     # 1
print(f"Cəm:         {np.sum(a)}")     # 15
```

## 8. Matplotlib — Vizuallaşdırma

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Line Plot — Xətt qrafiki
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)")
plt.xlabel("Zaman")
plt.ylabel("Amplituda")
plt.title("Sinus Dalğası")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Bar Chart — Sütunlu qrafik
növlər = ["Canavar", "Qartal", "Maral", "Ayı"]
saylar = [19, 11, 23, 5]

plt.figure(figsize=(8, 4))
plt.bar(növlər, saylar, color=["green", "gold", "brown", "gray"])
plt.title("Heyvan Müşahidələri")
plt.ylabel("Say")
plt.show()
```

```python
# Scatter Plot — Nöqtə qrafiki
x = np.random.randn(50)
y = 2 * x + np.random.randn(50) * 0.5

plt.figure(figsize=(6, 5))
plt.scatter(x, y, alpha=0.7)
plt.xlabel("X dəyişəni")
plt.ylabel("Y dəyişəni")
plt.title("Korrelyasiya")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.title("Normal Paylanma")
plt.xlabel("Dəyər")
plt.ylabel("Tezlik")
plt.show()
```
